데이터의 유형
- 정성적 데이터
    - 언어, 문자 등으로 기술된다.
- 정량적 데이터
    - 수치, 도형, 기호 등으로 기술된다.
        - 지역별 매출액, 영업이익률, 판매량 등.
    - 그 양이 크게 증가하더라도 이를 DBMS에 저장, 검색, 분석하여 활용하기가 용이하다.

암묵지와 형식지의 상호작용 관계
- 공통화
- 표출화
- 연결화
- 내면화

SQL 집계함수
- AVG: 지정한 열의 평균 값을 반환(수치형)
- COUNT: 테이블의 특정 조건이 맞는 것의 개수를 반환(수치형, 문자형)
- SUM: 지정한 열의 총합을 반환(수치형)
- STDDEV: 지정한 열의 분산을 반황(수치형)
- MIN: 지정한 열의 가장 작은 값을 반환(수치형)
- MAX: 지정한 열의 가장 큰 값을 반환(수치형)

데이터 마스킹
- 식과 같은 속성을 유지한 채, 새롭고 읽기 쉬운 데이터를 익명으로 생성하는 기술이다.

반정형 데이터
- 데이터 내부에 메타 데이터를 갖고 있으며 일반적으로 파일형태로 저장된다.

표출화
- 형식지 요소 중 하나.
- 개인에게 내재된 경험을 객관적인 데이터로 문서나 매체에 저장, 가공, 분석하는 과정이다.

DIKW 피라미드에서 개별 데이터 자체로는 의미가 중요하지 않은 객관적 사실에서 데이터를 가공 및 처리하여 정보, 지식, 지혜를 얻을 수 있다.

지식
- 상호 연결된 정보 패턴을 이해하여 이를 토대로 예측한 결과물이다.

Cinematch 시스템
- 넷플릭스(Netflix)에서 개발한 영화 추천 알고리즘이다

메타데이터
- 데이터에 대한 데이터로써 하위레벨의데이터를 설명/기술하려는 것이다.
>
인덱스
- 데이터베이스의 테이블에서 고속의 검색동작뿐만 아니라 레코드 접근과 관련 효율적인 순서 매김 동작에 대한 기초를 제공한다.

데이터 웨어하우스(Data Warehouse)
- 기업의 의사결정 과정을 지원하기 위한 주제 중심적으로 통합적이며 시간성을 가지는 비휘발성 데이터의 집합
>
데이터 웨어하우스의 특성
- 데이터의 주제 지향성
- 데이터 통합
- 데이터의 시계열성
- 데이터의 비휘발성

데이터 레이크(Data Lake)
- 별도로 정제되지 않은 자연스러운 상태의 아주 큰 데이터 세트이다.

데이터 마이닝(Data Mining)
- 대용량 데이터에서 의미있는 정보를 추출하여 의사결정에 활용하는 기술이다.
>
딥 러닝(Deep Learning)
- 다층구조 형태의 신경망을 바탕으로 하는 머신 러닝의 한 분야이다.

산업별 분석 애플리케이션의 분석 사례
- 에너지: 트레이딩, 공급/수요 예측 등

CRM
- 데이터베이스를 기초로 고객을 세부적으로 분류하여 효과적이고 효율적인 마케팅 전략을 개발한다.

ERP(Enterprise Resource Planning)
- 인사, 재무, 생산 등 기업의 전 부문에 걸쳐 독립적으로 운영되던 각종 관리 시스템의 경영자원을 하나의 통합 시스템으로 재구축함으로써 생산성을 극대화하려는 경영혁신기법을 의미한다.

사회기반 구조로서의 데이터베이스
- 물류, 지리/교통, 의료, 교육 등 부문에서 구축되었으며 활용이 되고 있다.

DIKW 단계
- 지식
    - 날씨가 따뜻해지고, 지점을 확장하여 올 8월 매출액은 3000만원으로 예상한다.
- 정보
    - 데이터의 가공 및 상관관계 간 이해를 통해 패턴을 인식하고 그 의미를 부여한 것이다.
    - 지식을 도출하기 위한 재료가 된다.
        - 지난 1년 매출액의 50%는 8월에 집중되어 있다.
        - 지난 1년 매출은 1월에서 8월까지 증가하였고, 12월까지 다시 증가하였다.
        - 8월 A상품 구매 고객의 80%가 40대 여성 고객으로 대부분 회사원이다.

빅데이터
- 대용량 데이터를 활용해 작은 용량에서는 얻을 수 없었던 새로운 통찰이나 가치를 추출해내는 일이다.
>
하둡
- 빅데이터 플랫폼 환경 궃축을 위해 사용할 뿐이다.
- 빅데이터가 하둡을 기반으로 하는 것은 아니다.

빅데이터 활용의 기본 3요소
- 데이터
- 기술
- 인력

빅데이터 출현 배경
- 고객데이터의 축적과 거대 데이터의 활용이 늘어남으로 필요한 기술 아키텍처 및 통계 도구들의 발전, 모바일 혁명 등의 관련기술의 발달을 들 수 있다.

빅데이터의 수집, 구축, 분석의 최종 목적
- 기존 방식으로는 얻을 수 없었던 통찰 및 가치 창출, 사업방식, 시장, 사회, 정부 등에서 변화와 혁신 주도이다.

플랫폼(Platform)
- 비즈니스 측면에서는 일반적으로 '공동 활용의 목적으로 구축된 유무형의 구조물'을 의미한다.
- 빅데이터가 최근에는 다양한 서드파티 비즈니스에 활용되면서 플랫폼 역할을 할 것으로 전망된다.
    - 페이스북 : 2006년 F8 행사를 기점으로 자신들의 소셜 그래프 자산을 외부 개발자들에게 공개하고 서드파티 개발자들이 페이스북 위에서 작동하는 앱을 만들기 시작하였다.
    - 하둡: 대규모 분산 병렬 처리의 업계 표준으로 맵리듀스 시스템과 분산 파일 시스템인 HDFS로 구성되며, 선형적인 성능과 용량 확장성, 고장 감내성을 가지고 있다.
    - 아마존(Amazon): S3와 BC2 환경을 제공함으로써 클라우드 서비스를최초로 실현하였다.

플랫폼형 비즈니스 모델
- 서비스, 기술 등의 기반 위에 다른 이해관계자들이 보완적인 상품, 서비스, 기술을 제공하는 생태계 구축을 목표로 하는 모델이다.

데이터 활용방법의 변화
- 빅데이터의 등장으로 데이터 수집비용의 감소와 클라우드 컴퓨팅 기술의 발전으로 데이터 처리비용이 감소하게 되었다.
- 이로 인해 표본을 조사하는 기존의 지식발견 방식에서 전수조사를 통해 샘플링이 주지 못하는 패턴이나 정보를 발견하는 방식으로 데이터 활용방법이 변화되었다.

데이터의 가치를 측정하기 어려운 이유
- 데이터 활용 방식: 재사용, 재조합(Mashup), 다목적용 개발
- 새로운 가치 창출
- 분석 기술 발전

데이터 난수화
- 고객의 과거 구매기록이나 나이, 수입, 건강정보와 같은 데이터가 해독이 불가능한 난수화를 통해 변경된 채로 기업에 전송된다.

소셜 네트워크 분석(Social Network Analysis)
- 사용자간의 소셜 관계를 알아내고자 할 때 이용하는 분석 방법이다.

유형 분석
- 문서를 분류하거나 조직을 그룹으로 나눌 때 또는 온라인 수강생들을 특성에 따라 분류할 때 사용하는 기법이다.
- 사용자가 어떤 특성을 가진 집단에 속하는 지 알아볼 때 사용한다.

연관성 분석
- 기업의 데이터베이스에서 상품의 구매, 서비스 등 일련의 거래 또는 사건들 간의 규칙을 발견하기 위한 분석이다.
- 흔히 장바구니 분석 등이 있다.

신용평가(Credit Rating)
- 투자자 보호를 위하여 금융상품 및 신용공여 등에 대하여 그 원리금이 상환될 가능성과 기업, 법인 및 간접투자기구 등의 신용도를 평가하는 행위이다.
- 핀테크 분야에서 빅데이터 활용이 활발하게 이루어지고 있다.

빅데이터 시대에 발생할 수 있는 위기 요인과 해결 방법
- 사생활 침해
    - 정보 사용자 책임제로 변환.
- 책임 원칙 훼손
    - 결과기반 책임 원칙 고수.
- 데이터 오용
    - 알고리즘 접근 허용.

빅데이터 시대의 위기와 통제
- 빅데이터 분석은 일어난 일에 대한 데이터에 의존하므로 예측의 정확도는 높지만 항상 맞을 수는 없어 데이터 오용의 피해가 발생할 수 있다.
- 빅데이터가 발생시키는 문제를 중간자 입장에서 중재하며 해결해 주는 알고리즈미스트(Algorithmist)도 새로운 직업으로 부상하게 될 것이다.
- 개인정보 사용자의 정보사용에 대한 무한책임의 한계로 개인정보 사용 동의제보다 책임제로 더욱 강화시켜야 한다.
- 민주주의 국가의형사 처벌과 같이 잠재적 위협이 아닌 명확하게 행동한 결과에 대해 책임을 묻기 때문에 빅데이터 사전 성향 분석을 실시한다면 책임 원칙을 훼손한다.

사물인터넷(Internet of Things)
- 인터넷 기반으로 모든 사물을 연결해 사람과 사물, 사물과 사물 간의 정보를 상호 소통하는 지능형 기술 및 서비스이다.
- 사물에서 생성되는 Data를 활용한 분석을 통해 마케팅 등에 활용할 수 있다.

객체지향 DB

- 멀티미디어 등 복잡한 데이터 구조를 관리하는 DBMS이다.

SCM(Supply Chain Management)
- 기업이 외부 공급업체 또는 제휴업체와 통합된 정보시스템으로 연계하여 시간과 비용을 최적화시키기 위한 것이다.
- 자재 구매, 생산, 제고, 유통, 판매, 고객 데이터로 구성된다.

유전자 알고리즘(Genetic Algorithms)
- 생명의 진화를 모방하여 최적해(Optimal Solution)를 구하는 알고리즘으로 존 홀랜드(John Holland)가 1975년에 개발하였다.
- '최대의 시청률을 얻으려면 어떤 시간대에 방송해야 하는가?'와 같은 문제를 해결할 때 사용된다.
- 어떤 미지의 함수 Y=f(x)를 최적화하는 해 x를 찾기 위해, 진화를 모방한(Simulated Evolution) 탐색 알고리즘이라고 말할 수 있다.
